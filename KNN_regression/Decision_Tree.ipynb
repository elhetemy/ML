# Feature encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data[['MAKE','MODEL','VEHICLECLASS','TRANSMISSION','FUELTYPE']]=data[['MAKE','MODEL','VEHICLECLASS','TRANSMISSION','FUELTYPE']].apply(le.fit_transform)
data.head()
```

```python
data.info()
```

# data splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['CO2EMISSIONS'],axis=1).values
y=data['CO2EMISSIONS'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# applying DecisionTree Model

```python
from sklearn.tree import DecisionTreeRegressor
D_T_reg_model=DecisionTreeRegressor(random_state=42)
D_T_reg_model.fit(x_train,y_train)
y_pred=D_T_reg_model.predict(x_test)
```

# Model evaluation

```python
from mlxtend.evaluate import bias_variance_decomp
avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(D_T_reg_model,x_train,y_train,x_test,y_test, loss='mse',random_seed=42)
print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

```python
from sklearn import tree

columns_names = data.drop('CO2EMISSIONS',axis=1).columns.tolist()
columns_names

print(tree.export_text(D_T_reg_model, feature_names=columns_names))
```

```python
import graphviz

tree_fig = tree.export_graphviz(D_T_reg_model,feature_names=columns_names,filled=True)

graph = graphviz.Source(tree_fig,format='png')
graph.render('tree')
```

```python
import matplotlib.pyplot as plt

importance = D_T_reg_model.feature_importances_

indices = np.argsort(importance)

plt.barh(range(len(indices)),importance[indices], color='b')

plt.yticks(range(len(indices)), [columns_names[i] for i in indices])
```
